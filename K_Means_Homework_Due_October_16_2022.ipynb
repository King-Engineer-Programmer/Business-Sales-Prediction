```python

#Learning Agent: kNN
#Semi Autonomous

#Load necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# # This code will group men and women based on how much money they which assocaites them to a customer persona

def convert(value):
    if value=="Group Two Customer":
        return 2
    elif value=="Group One Customer":
        return 1
    elif value=="Group Three Customer":
        return 3
    elif value=="Group Four Customer":
        return 4
    

class Agent_kNN:   
    #Load the environment - dataset            
    def load_Data(self, worldpath):         
        # Read dataset to pandas dataframe
        data = pd.read_csv(worldpath)
        data=data.dropna()
        data.info()
        
        #convert label to numeric for this framework
        print("\n Customer Class Raw data unique:", data['Persona'].unique())
        #le = LabelEncoder()
        #data['Persona']= le.fit_transform(data['Persona']) 
        data['Persona']=data['Persona'].apply(convert)
        #print("Customer codes: ", data['Persona'].unique())
        
        #Preprocessing: extract independent-x and dependent-y variables        
        y_data = data['Persona']
        X_data = data.drop('Persona', axis =1)  
        #print('\n', data.head())      
                               
        #Training and Test Set Data Split
        X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.10, random_state=0)
        self.X_train = X_train 
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test       
                             
    def train_agent_predict_test(self):         #Learning and Predicting evaluation test set
        kNN_learner= KNeighborsClassifier(n_neighbors=5) #Hyperparameter
        kNN_learner.fit(self.X_train, self.y_train)
        self.kNN_learner = kNN_learner
        
        #predict test data
        self.pred_y = kNN_learner.predict(self.X_test)
        
    def evaluate_kNN_learner(self):    
        print('\n')
        print(classification_report(self.y_test, self.pred_y))
        print(confusion_matrix(self.y_test, self.pred_y))
        
    def predict_new_percept(self, new_percept):                                                       
        #predict and display meaningful result
        new_pred = self.kNN_learner.predict(new_percept)
        print('\n Predicted customer group: = ', new_pred) 
        
        if new_pred == 1:
            print("Group One Customer")
        elif new_pred == 2:
            print("Group Two Customer")
        elif new_pred == 3:
            print("Group Three Customer")
        elif new_pred == 4:
            print("Group Four Customer")
        else:
            print("Unknown flower")  
              
        
        
        
        
#--------------------------class driver--------------------------------------------------------------
handle = Agent_kNN()
worldpath = 'Persona_Matching_3.csv'                   #initial percepts

handle.load_Data(worldpath)
handle.train_agent_predict_test()
handle.evaluate_kNN_learner()

# Real-world and new percept of iris flowers  
# Learning Agent designed as semi-autonomous 
print('\n Start Predicating Customer Personas \n')
while True:    
    n = input("Please press Enter-key to continue or enter 'q' to Quit: ")
    if n.strip() == 'q':
        break
    
    #enter new features, format like X_data          
    print("Percept of New customer description to be predicted: ")
    Age = float(input("Enter age: "))
    Gender = float(input("Enter gender if Male Enter 1 if Female enter 0: "))
    Salary = float(input("Enter your salary with no commas: "))
    
    
    #Prepare the 2D List data structure for authentication   
    new_customer = []
    new_customer.append([Age, Gender, Salary])
    print('\n Structured new customer description: ', new_customer)     
    handle.predict_new_percept(new_customer)
```
