```python


#Load necessary libraries
import pandas as das  
import numpy as np  
import matplotlib.pyplot as vis  
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics



#df = pd.read_csv('advertising.csv')

#Main program built as OOP -Object Oriented Programming
class Learning_agent:
     #Load the world of Stock exchange returns - dataset            
     def load_Data(self, worldpath, cols):       
        data = das.read_csv(worldpath, names=cols)
                     
         #Preprocessing: independent-x and dependent-y variables
        X_data = data[[cols[0], cols[1], cols[2]]].values
        y_data = data[cols[3]].values
                        
         #Training and Test Set Data Split
        X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.20, random_state=0)
        self.X_train = X_train 
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test   
             
     def train_agent_predict_test(self):         
        MLR_agent = LinearRegression()  
        MLR_agent.fit(self.X_train, self.y_train)
        self.MLR_agent = MLR_agent   
        self.pred_y = MLR_agent.predict(self.X_test)
        
     def evaluate_learning_agent(self):      
          #visualization of predicted and actual values
        main_pred = das.DataFrame({'Actual Sale': self.y_test, 'Predicted Sale': self.pred_y})
        main_pred_sub = main_pred.head(15)
        print(main_pred_sub)
        
         #Beautify the graphs
        main_pred_sub.plot(kind='line', marker='o', linewidth = 2, figsize=(15,10))
        vis.grid(which='major', linestyle=':', linewidth='0.6', color='green')
        vis.grid(which='minor', linestyle='-', linewidth='0.5', color='red')
        vis.show()

#          #utility functions: performance of the learning agent
        print('\n Mean Absolute Error= ', metrics.mean_absolute_error(self.y_test, self.pred_y))  
        print('\n Mean Squared Error= ', metrics.mean_squared_error(self.y_test, self.pred_y))  
        print('\n Root Mean Squared Error= ', np.sqrt(metrics.mean_squared_error(self.y_test, self.pred_y)))
        
     def new_percepts(self, new_instance):                                                
        pred_stock = self.MLR_agent.predict(new_instance)
        print('\n Predicted Istanbul_Stock_Exchange_returns: = ', pred_stock)    
        




# #----------------------------class driver-------------------------------------------------
handle = Learning_agent()
worldpath = 'advertising2.csv'
# #worldpath = 'stock_exchange_regression2.csv'

# #Data headings and calling methods
cols = ['TV', 'Radio', 'Newspaper', 'Sales']
handle.load_Data(worldpath, cols)
handle.train_agent_predict_test()       
handle.evaluate_learning_agent()        
        

        
        
        
```
